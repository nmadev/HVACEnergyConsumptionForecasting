```python
import matplotlib.pyplot as plt 
import seaborn as sns 
import numpy as np
```

```python
data = [90, 6, 4] 
keys = ['Non-HVAC', 'Residential HVAC', 'Commercial HVAC'] 
  
explode = [0, 0.15, 0] 
# palette_color = sns.color_palette('light') 

# plt.title("Energy Consumption in the United States")
plt.pie(data, labels=keys, 
        explode=explode, autopct='%.0f%%') 
plt.savefig('./final_figures/MISC_figures/ECPie.png', dpi=200, bbox_inches='tight')
plt.show()
```

```python
def sinusoid(tmax, tmin, x):
    return (tmax - tmin) / 2. * np.cos(2. * np.pi * x / 365 + np.pi) + (tmax + tmin) / 2.

tmin = 12
tmax = 23
tac = 18.3

x_short = np.arange(95, 260, 0.1)

plt.figsize
x = np.arange(0, 365, 0.1)
plt.plot(x, sinusoid(tmax, tmin, x), color='black', label='T(D)')
plt.plot(x, [tac] * len(x), color='black', linestyle='--', label="T_AC")
plt.fill_between(x_short, sinusoid(tmax, tmin, x_short), [tac] * len(x_short), hatch='/', alpha=0)
plt.title("Example Annual Temperature Curve")
plt.xlabel('Day of the Year')
plt.ylabel('Temperature (C)')
plt.legend()
plt.savefig('./final_figures/MISC_figures/AnnualTemperatureCurve.png', dpi=200, bbox_inches='tight')
```
